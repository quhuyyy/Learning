```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("/content/drive/MyDrive/Python/data_final.xlsx")
data
```

```python
data2 = data.copy()
data2 = data2[data['ID number'] == 2212943]
data2
```

```python
data2 = data2.iloc[:, 2::3]
data2
```

```python
len(data2.columns)
```

```python
new_columns = []
for column in data2.columns:
  column = column.split("_")
  new_column = column[0].split("lab")
  new_column[0] = new_column[0][1:3]
  final_column = "".join(new_column)
  new_columns.append(final_column)
print(len(new_columns))
new_columns
```

```python
data2.columns = new_columns
data2
```

```python
data2 = data2.fillna(0)
data2
```

```python
def show_score(data2, typeLab = "Prelab", color = 'pink'):
  if typeLab == 'Prelab':
    x = [0, 5, 14, 18, 26, 28]
  elif typeLab == 'Inlab':
    x = [5, 10, 18, 22, 28, 30]
  elif typeLab == 'Postlab':
    x = [10, 14, 22, 26, 30, 33]
  else:
    raise NameError("Không tồn tại lab")

  new_columns = [col[2:] for col in data2.columns]

  # print(data2.columns)

  plt.figure(figsize=(20,10))

  plt.subplot(2, 2, 1)
  plt.plot(new_columns[x[0]:x[1]], data2.iloc[0, x[0]:x[1]],marker='o', color=color)
  plt.xlabel("Tên chủ đề")
  plt.yticks(np.linspace(0, 10, 11))
  plt.ylabel("Điểm")
  plt.title(typeLab+"1")

  plt.subplot(2, 2, 2)
  plt.plot(new_columns[x[2]:x[3]], data2.iloc[0, x[2]:x[3]],marker='o', color=color)
  plt.xlabel("Tên chủ đề")
  plt.yticks(np.linspace(0, 10, 11))
  plt.ylabel("Điểm")
  plt.title(typeLab+"2")

  plt.subplot(2, 2, 3)
  plt.plot(new_columns[x[4]:x[5]], data2.iloc[0, x[4]:x[5]],marker='o', color=color)
  plt.xlabel("Tên chủ đề")
  plt.yticks(np.linspace(0, 10, 11))
  plt.ylabel("Điểm")
  plt.title(typeLab+"3")
```

```python
show_score(data2, color= "cyan")
```

```python
show_score(data2, "Inlab", "pink")
```

```python
show_score(data2,"Postlab", color= "violet")
```

```python
x = [0, 5, 10, 14, 18, 22, 26, 28, 30, 33]

new_columns = [col[2:] for col in data2.columns]

# new_columns = data2.columns

# print(data2.columns)
plt.figure(figsize=(20,10))

for i in range(0, 9, 3):
  plt.subplot(2, 2, (i//3)+1)
  plt.plot(new_columns[x[i]:x[i+1]], data2.iloc[0, x[i]:x[i+1]],marker='o', linewidth=2, linestyle='--', color='cyan', label = "Prelab")
  plt.plot(new_columns[x[i]:x[i+1]], data2.iloc[0, x[i+1]:x[i+2]],marker='o', linewidth=1.5, linestyle='--', color='pink', label = "Inlab")
  if i > 0:
    plt.plot(new_columns[x[i]:x[i+1]], data2.iloc[0, x[i+2]:x[i+3]],marker='o', linewidth=1, linestyle='--', color='violet', label = "Postlab")
  plt.xlabel("Tên chủ đề")
  plt.yticks(np.linspace(0, 10, 11))
  plt.ylabel("Điểm")
  plt.title("Lab1")
  plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

```python
data_prelab = data2.iloc[0, [0,1,2,3,4,14,15,16,17,26,27]]
data_prelab
```

```python
data_inlab = data2.iloc[0, [5,6,7,8,9,18,19,20,21,28,29]]
data_inlab
```

```python
data_postlab = data2.iloc[0, [10,11,12,13,22,23,24,25,30,31]]
data_postlab
```

```python
prelab_with_10 = data_prelab[data_prelab == 10]
prelab_without_10 = data_prelab[data_prelab != 10]
count_10_prelab = [len(prelab_with_10), len(prelab_without_10)]

inlab_with_10 = data_inlab[data_inlab == 10]
inlab_without_10 = data_inlab[data_inlab != 10]
count_10_inlab = [len(inlab_with_10), len(inlab_without_10)]

postlab_with_10 = data_postlab[data_postlab == 10]
postlab_without_10 = data_postlab[data_postlab != 10]
count_10_postlab = [len(postlab_with_10), len(postlab_without_10)]
```

```python
plt.figure(figsize=(15,5))

plt.subplot(2, 2, 1)
plt.pie(count_10_prelab, autopct = "%.2f", colors = ['#00CCDD', '#7CF5FF'])
plt.title("Prelab")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1), labels=['Số bài 10 điểm', 'Số bài khác 10 điểm'])

plt.subplot(2, 2, 2)
plt.pie(count_10_inlab, autopct = "%.2f", colors = ['#00CCDD', '#7CF5FF'])
plt.title("Inlab")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1), labels=['Số bài 10 điểm', 'Số bài khác 10 điểm'])

plt.subplot(2, 2, 3)
plt.pie(count_10_postlab, autopct = "%.2f", colors = ['#00CCDD', '#7CF5FF'])
plt.title("Postlab")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1), labels=['Số bài 10 điểm', 'Số bài khác 10 điểm'])

plt.tight_layout()
```
